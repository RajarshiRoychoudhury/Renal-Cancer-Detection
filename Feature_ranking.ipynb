```python
for columns in df.columns:
    X=df[columns].unique()
    map_dictionary={}
    count=1
    for j in range(len(X)):
        if X[j] not in map_dictionary :
            if not str(X[j]).isnumeric():
                map_dictionary[X[j]]=count
                count+=1
            else:
                map_dictionary[X[j]]=X[j]
            
    #print(map_dictionary)
    for j in range(len(df[columns])):
        df.at[j,columns]=map_dictionary[df[columns][j]]
```

```python
y=df.tumor_grade.values
```

```python
y=y.astype('int')
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
df.drop('tumor_grade',axis=1,inplace=True)
```

```python
def model_evaluate(X,y):
    X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
    clf=svm.LinearSVC(multi_class='crammer_singer')
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    return ((y_test==y_pred).sum())/len(X)
    
    
```

```python
def make_dict(df,y):
    acc_dict={}
    for columns in df.columns:
        l=df[columns].values
        X=[]
        max=0
        for j in range(len(l)):
            X.append([l[j]])
            if l[j]>max:
                max=l[j]
        for j in range(len(X)):
            X[j][0]=(X[j][0]*1.0)/max
        acc_dict[columns]=model_evaluate(X,y)
    return acc_dict
```

```python
dict=make_dict(df,y)
print(dict)
```

```python
l=[]
for j in dict:
    l.append(dict[j])
l.sort()
l.reverse()
print(l)
```

```python
l2=[]
def get_key(val): 
    for key, value in dict.items(): 
         if val == value and key not in l2: 
                l2.append(key)
                return key 
for i in l:
    print(get_key(i))
```
